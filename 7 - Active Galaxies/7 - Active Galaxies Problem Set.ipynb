```python
from astropy.io import fits

# The GAMA spectroscopic line strength catalog.
hdul = fits.open('GaussFitSimple.fits')

print(hdul[1].columns)
data = hdul[1].data

# identify all the line strengths needed.
OI = data['OIR_FLUX']
OII = data['OIIB_FLUX']
OIII = data['OIIIB_FLUX']
NII = data['NIIR_FLUX']
Halpha = data['HA_FLUX']
Hbeta = data['HB_FLUX']
SII = data['SIIB_FLUX']

# define the signal-to-noise of each line
OI_sn   = data['OIR_FLUX']/data['OIR_FLUX_ERR']
OII_sn   = data['OIIB_FLUX']/data['OIIB_FLUX_ERR']
OIII_sn   = data['OIIIB_FLUX']/data['OIIIB_FLUX_ERR']
NII_sn    = data['NIIR_FLUX']/data['NIIR_FLUX_ERR']
Halpha_sn = data['HA_FLUX']/data['HA_FLUX_ERR']
Hbeta_sn  = data['HB_FLUX']/data['HB_FLUX_ERR']
SII_sn    = data['SIIB_FLUX']/data['SIIB_FLUX_ERR']


# signal-to-noise cut.
sn = np.where( (OI_sn>3)&(OII_sn>3)&(OIII_sn>3)&(NII_sn>3)&(SII_sn>3)&(Halpha_sn>3)&(Hbeta_sn>3))

# line strengths
OI = OI[sn]
OII = OII[sn]
OIII = OIII[sn]
NII = NII[sn]
SII = SII[sn]
Halpha = Halpha[sn]
Hbeta = Hbeta[sn]

```

## BPT diagram

First we load in the Emission Line strengths from GAMA. We need OIII, NII, H$\alpha$ and H$\beta$. 
Several authors defined a curve between the ratios of line strengths to identify AGN-dominated and star-formation.

Kauffmann+ (2003) criterion:

$$ log([OIII]/H\beta) > 0.61/(log([NII]/H\alpha) − 0.05) + 1.3 $$
