```python
import requests
import IPython.display
import numpy as np
from matplotlib import pyplot as plt
from astropy import units as u
from astropy.coordinates import SkyCoord
from astropy.table import Table

impix = 1024
imsize = 12*u.arcmin

import random
ra = random.uniform(0, 360.)
dec = random.uniform(0, 90.)

print(ra,dec)
#hcg7_center = SkyCoord(9.81625*u.deg, 0.88806*u.deg, frame='icrs')
hcg7_center = SkyCoord(ra*u.deg, dec*u.deg, frame='icrs')



cutoutbaseurl = 'http://skyservice.pha.jhu.edu/DR12/ImgCutout/getjpeg.aspx'

req = requests.get(cutoutbaseurl, dict(ra=hcg7_center.ra.deg, 
                                       dec=hcg7_center.dec.deg, 
                                       width=impix, height=impix, 
                                       scale=imsize.to(u.arcsec).value/impix))
with open('SDSS_cutout.jpg', 'wb') as f:
    f.write(req.content)
IPython.display.Image('SDSS_cutout.jpg')
```
